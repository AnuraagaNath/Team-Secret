```python
# imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lazypredict.Supervised import LazyRegressor
```

```python
# read data
train = pd.read_parquet('train.parquet')
train.head()
```

```python
train.info()
```

```python
# dependent and independent variable
X = train.drop(['FloodProbability'], axis=1)
y = train['FloodProbability']
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# model training and evaluation of r2 score
def train(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = r2_score(y_test, y_pred)
    return score
```

```python
# model initializations
linear = LinearRegression()
xgb = XGBRegressor()
cat = CatBoostRegressor(verbose=0)
```

```python
# evaluation of each model without hyper parameter tuning
linear_score = train(linear)
xgb_score = train(xgb)
cat_score = train(cat)
```

```python
scoreboard = pd.DataFrame()
scoreboard['Linear Regression'] = [linear_score]
scoreboard['XGBoost Regression'] = [xgb_score]
scoreboard['Catboost Regression'] = [cat_score]
scoreboard
```

```python
lazypredict = LazyRegressor()
models, predicts = lazypredict.fit(X_train, X_test, y_train, y_test)
```
